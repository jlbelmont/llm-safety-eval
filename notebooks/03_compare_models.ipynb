# 03 - Compare Models

Minimal scaffold to compare label distributions across models.

```python
import pandas as pd

labels_path = '../runs/example_run/labels.jsonl'
try:
    df = pd.read_json(labels_path, lines=True)
    summary = df.groupby(['model_id', 'label']).size().unstack(fill_value=0)
    summary
except FileNotFoundError:
    print('No labels file found yet; run labeling CLI to generate labels.jsonl')
```
